```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
```

```python
data_set = pd.read_csv('Dataset/voice.csv')
data_set
```

```python
X = data_set.drop(columns=['label', 'meandom', 'mindom', 'maxdom', 'dfrange', 'modindx'])
X
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=15)  
X_pca = pca.fit_transform(X_scaled)
```

```python
k_means = KMeans(n_clusters=3, random_state=0)
k_means.fit(X_pca)
```

```python
data_set['cluster'] = k_means.labels_
```

```python
cluster_map = {
    0: 'female',  
    1: 'male'     
}
```

```python
data_set['predicted_label'] = data_set['cluster'].map(cluster_map)
```

```python
data_set['true_label'] = data_set['label']
```

```python
correct_labels = data_set['predicted_label'] == data_set['true_label']
accuracy = correct_labels.mean()
print(f"Clustering accuracy: {accuracy * 100:.2f}%")
```
